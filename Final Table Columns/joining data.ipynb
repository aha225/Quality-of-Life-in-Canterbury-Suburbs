```r
library(tidyverse)
library(singer)
library(skimr)
```

```r
crash_df <- read.csv("crash_final.csv")
crime_df <- read.csv("crime_by_area.csv")
hospital_df <- read.csv("Closest_Hospitals.csv")
school_df <- read.csv("Area Unit Schools.csv")

names(crash_df)[1] <- "AreaUnit" # renaming column name to be same for joining later
names(hospital_df)[1] <- "AreaUnit"
names(school_df)[2] <- "AreaUnit"
school_df2 = subset(school_df, select = -1 ) # removing first column which is just id numbers
school_df2
```

```r
crash_and_crime<-crash_df %>% full_join(crime_df, by="AreaUnit")
```

```r
c_c_h_df<-crash_and_crime %>% full_join(hospital_df, by="AreaUnit")
```

```r
final_df <- c_c_h_df %>% full_join(school_df2, by="AreaUnit")
final_df
final_df$Primary[is.na(final_df$Primary)] <- 0
select(final_df, AreaUnit, Total_Crashes, Total_Crime, Closest_Medical_Hospital, Average_decile, Primary)
```

```r
write.csv(final_df,"Quality of life in Canterbury suburbs.csv", row.names = FALSE)
```
