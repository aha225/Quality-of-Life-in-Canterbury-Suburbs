DATA422 GROUP PROJECT SCHOOLS DIRECTORY PROCESSING


```r
# Loading libaries and packages 
#install.packages("RCurl")
library(tidyverse)
library(readr)

# Load directory
school_directory <- read.csv("SchoolsDirectory.csv")

# Show directory
school_directory
```

```r
# Filtering database for schools only in Canterbury and Chatham Islands
canterbury_school_directory = school_directory %>% 
filter(Education.Region == 'Canterbury/Chatham Islands')
```

```r
# Checking how many rows are in the canturbury school table
nrow(canterbury_school_directory)

# Looking at column names and more database features
library(skimr)
canterbury_school_directory %>% 

skim()
```

```r
# Renaming columns for better names

canterbury_school_directory <- canterbury_school_directory %>% 
rename(School_name = School.Name,
       Email = Email., 
       Principal = Principal., 
       School_website = School.Website, 
       City = Town...City, 
       Postal_address = Postal.Address, 
       Postal_address_suburb = Postal.Address.Suburb, 
       Postal_address_city = Postal.Address.City,
       Urban_area = Urban.Area,
       School_type = School.Type,
       Student_gender = Gender.of.Students,
       Regional_council= Regional.Council,
       Ministry_of_Education_local_office = Ministry.of.Education.Local.Office,
       Education_Region = Education.Region,
       General_electorate= General.Electorate,
       Maori_electorate = M.U.0101.ori.Electorate,
       Area_unit = Area.Unit,
       Community_of_learning_name = Community.of.Learning.Name,
       Isolation_index = Isolation.Index,
       School_number = X.U.FEFF.School.Number,
       Postal_code = Postal.Code,
       Community_of_learning_id = Community.of.Learning.ID,
       European_Pakeha_enrolled = European...P.U.0101.keh.U.0101.,
       Maori_enrolled = M.U.0101.ori,
       Total_school_roll = Total.School.Roll
       )       
       
canterbury_school_directory
```

```r
# Setting empty strings to NA
canterbury_school_directory[canterbury_school_directory==""]<-NA

# Setting suburb equal to city if suburb does not exist due to town being outside of christchurch
canterbury_school_directory$Suburb[is.na(canterbury_school_directory$Suburb)] <- canterbury_school_directory$City[is.na(canterbury_school_directory$Suburb)]

# Remove Schools with decile greater than 10 and thus invalid
canterbury_school_directory = canterbury_school_directory %>% 
filter(!Decile > 10)
```

```r
# Attaching library
library(dplyr)

# Grouping by area unit and calculating average decile per area unit
area_unit_schools1 = canterbury_school_directory %>% 
    group_by(Area_unit) %>% 
            summarise(Average_decile = mean(Decile))

# Grouping by area unit and school type, and getting a count for each type of school eithin area unit - long form dataframe
area_unit_schools2 = canterbury_school_directory %>% 
    group_by(Area_unit, School_type) %>% 
                      tally()

# Convert to wide format to get number of schools for each type as a column per area unit row
area_unit_schools2 = area_unit_schools2 %>% 
spread(key = School_type, value = n)  %>% 
mutate_all(funs(ifelse(is.na(.), 0, .))) # Setting = 0 if NA
area_unit_schools2
```

```r
# Now we join these up so we have a final table with all the columns we want
area_unit_schools = inner_join(area_unit_schools1, area_unit_schools2, by = 'Area_unit')
area_unit_schools
```

```r
# Next we group all types of school columns into 4 different - primary, secondary, composite and other
area_unit_schools = area_unit_schools %>% 
rename("Full_Primary" = "Full Primary",                 # renaming columns to eliminate whitespace
       "Secondary1"  = "Secondary (Year 7-15)",     
       "Secondary2"  = "Secondary (Year 9-15)",
       "Special_School" = "Special School",
       "Teen_Parent_Unit" = "Teen Parent Unit") %>% 
mutate(Primary = Contributing + Full_Primary,           # Summing relevant columns
       Secondary = Secondary1 + Secondary2,
       Other = Intermediate + Special_School + Teen_Parent_Unit) %>% 
select(Area_unit, Average_decile, Primary, Secondary, Composite, Other)  # selecting relevant columns
 
```

```r
# Saving files
write.csv(canterbury_school_directory, 'Canterbury School Directory')
write.csv(area_unit_schools, 'Area Unit Schools')
```

```r
# See some insights - show all area units with more than one school and rank by highest average decile
area_unit_schools %>%
arrange(desc(Average_decile)) %>% 
filter(Primary + Secondary + Composite + Other > 1)
```
