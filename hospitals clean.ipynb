```r
public.df <- read.csv("LegalEntitySummaryPublicHospital.csv")
private.df <- read.csv("LegalEntitySummaryNGOHospital.csv")
head(public.df)
```


Combine the public and private data sets.

```r
hospitals.df <- rbind(public.df, private.df)
```

Filter for the hostitals in the Canterbury region. THis was done using the district health boards. Note that South Canterbury DHB needs to be included to include Timaru.

```r
hospitals_Can.df <- hospitals.df %>% filter(DHB.Name == "Canterbury District Health Board" | DHB.Name == "South Canterbury District Health Board")
```

The columns that gave information that was not relavant for the overall dataset were removed. This primarily included the data about location of the legal entity of the hospital, which is not important.

```r
hospitals_Can1.df <- hospitals_Can.df %>% select(Premises.Name:DHB.Name) %>% select(-Premises.Website) %>% select(-Premises.Address.Other)
hospitals_Can1.df
```

```r
write_csv(hospitals_Can1.df, "hospitals_Can1.df.csv") # Save the bike_df dataframe.
```
